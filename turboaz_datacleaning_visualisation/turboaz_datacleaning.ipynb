```python
cols_for_outliers=['price_azn', 'kilometrage_num', 'at_gucu', 'engine_displacement_num']
```

```python
q1 = df1[cols_for_outliers].quantile(0.25)
q3 = df1[cols_for_outliers].quantile(0.75)
iqr = q3 - q1
upper_limit = q3 + iqr * 5
lower_limit = q1 - iqr * 5
```

```python
upper_limit,lower_limit
```

```python
upper_outlier = df1[cols_for_outliers] > upper_limit
lower_outlier = df1[cols_for_outliers] < 0
outliers = upper_outlier.any(axis=1) | lower_outlier.any(axis=1)
```

```python
df1.loc[outliers, :].shape
```

```python
df1.loc[~outliers, cols_for_outliers].hist(bins=50, figsize=(10,12));
```
